### 6. Correla√ß√µes

```python
# Para valores num√©ricos
# Matriz de correla√ß√£o
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Mapa de Correla√ß√£o')
plt.show()
```

```python
#confirme a correlacao entre sleephours e stresslevel usando spearman e pearson usando o coefinciente e o p-valor
from scipy.stats import pearsonr, spearmanr
pearson_corr, pearson_p = pearsonr(df['SleepHours'], df['StressLevel'])
spearman_corr, spearman_p = spearmanr(df['SleepHours'], df['StressLevel'])
print(f'Pearson correlation: {pearson_corr}, p-value: {pearson_p}')
print(f'Spearman correlation: {spearman_corr}, p-value: {spearman_p}')
```

```python
#confirma a correlacao entre genero e colesterol usando chi-square test
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['Gender'], df['Cholesterol'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f'Chi-square statistic: {chi2}, p-value: {p}')
```

### 7. Rela√ß√µes com Doen√ßa Card√≠aca

```python
sns.boxplot(x='HeartDisease', y='BMI', data=df)
plt.title('Distribui√ß√£o do IMC por presen√ßa de Doen√ßa Card√≠aca')
plt.show()
```

```python
#cofirmar se ha relacao entre IMC e Doen√ßa Card√≠aca usando t-test
contingency_table = pd.crosstab(df['BMI'], df['HeartDisease'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f'Chi-square statistic: {chi2}, p-value: {p}')
```
