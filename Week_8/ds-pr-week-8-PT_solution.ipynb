### 4. Quais s√£o as vari√°veis num√©ricas que apresentam maior dispers√£o (diferen√ßa entre Q3 e Q1)?
> - Dica: utiliza `df.describe()`ou calcula o IQR (*Interquartile Range*).
> - Explique por que uma alta dispers√£o pode ser relevante em uma an√°lise de pre√ßos.

```python
#Quais vari√°veis num√©ricas apresentam maior dispers√£o (diferen√ßa entre Q3 e Q1)?
numeric_cols = df.select_dtypes(include=[np.number]).columns
dispersion = {}
for col in numeric_cols:
    if col == 'id':  # ignore id column
        continue
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    #check if IQR is nan
    if IQR:
        dispersion[col] = IQR

# # Ordenar por dispers√£o
# dispersion = dict(sorted(dispersion.items(), key=lambda item: item[1], reverse=True))
# print("Dispers√£o (IQR) das vari√°veis num√©ricas:")
# for col, iqr in dispersion.items():
#     print(f"{col}: {iqr}")

#print a proportion of dispersion to the mean of each column
print("\nPropor√ß√£o de dispers√£o para a m√©dia:")
for col, iqr in dispersion.items():
    mean = df[col].mean()
    proportion = iqr / mean if mean != 0 else 0
    print(f"{col}: {proportion:.2f}")
```

```python
# Mostrar o histograma das 5 vari√°veis com maior dispers√£o
top_5_disp = list(dispersion.keys())[:5]
for col in top_5_disp:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribui√ß√£o de {col}')
    plt.show()
```

```python
# Imprimir o intervalo de valores da vari√°vel "bathrooms", que √© a mais dispersa, usando describe
df['bathrooms'].describe()
```

```python
# IQR significa:
# O intervalo interquartil (IQR) √© uma medida de dispers√£o estat√≠stica, ou seja, indica qu√£o dispersos est√£o os valores num conjunto de dados.
# √â calculado como a diferen√ßa entre o terceiro quartil (Q3) e o primeiro quartil (Q1) dos dados.
# Uma elevada dispers√£o (IQR) no pre√ßo indica uma grande variedade de pre√ßos das casas, o que pode sugerir um mercado diversificado com diferentes tipos de propriedades.
# Isto √© relevante para compradores e vendedores, pois reflete a variabilidade nos valores das propriedades, ajudando-os a tomar decis√µes informadas com base nas tend√™ncias do mercado e nas caracter√≠sticas dos im√≥veis.
```

### 6. H√° outliers no pre√ßo? Mostrar as linhas do dataframe com os outliers.

```python
# Existem valores discrepantes (outliers) nos pre√ßos?
# Imprimir as linhas do dataframe que cont√™m outliers
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['price'] < (Q1 - 1.5 * IQR)) | (df['price'] > (Q3 + 1.5 * IQR))]
print("Outliers em pre√ßos:")
# print(outliers)
print(f"{len(outliers)} of {len(df)} rows are outliers in price.")
```

```python
# Utiliz√°mos (Q1 - 1.5 * IQR) como limite para os outliers inferiores e (Q3 + 1.5 * IQR) para os outliers superiores.
# Isto significa que qualquer pre√ßo abaixo de Q1 - 1.5 * IQR ou acima de Q3 + 1.5 * IQR √© considerado um outlier.
# Propor√ß√£o de outliers nos pre√ßos
proportion_outliers = len(outliers) / len(df) * 100
print(f"Propor√ß√£o de outliers em pre√ßos: {proportion_outliers:.2f}%")
```

### 11 ‚Äì Identifique poss√≠veis outliers em price e sqft_living. Esses valores devem ser removidos ou mantidos? Justifique.

```python
#Identifique poss√≠veis outliers em price e sqft_living. Esses valores devem ser removidos ou mantidos? Justifique.
Q1_price = df['price'].quantile(0.25)
Q3_price = df['price'].quantile(0.75)
IQR_price = Q3_price - Q1_price
outliers_price = df[(df['price'] < (Q1_price - 1.5 * IQR_price)) | (df['price'] > (Q3_price + 1.5 * IQR_price))]
Q1_sqft = df['sqft_living'].quantile(0.25)
Q3_sqft = df['sqft_living'].quantile(0.75)
IQR_sqft = Q3_sqft - Q1_sqft
outliers_sqft = df[(df['sqft_living'] < (Q1_sqft - 1.5 * IQR_sqft)) | (df['sqft_living'] > (Q3_sqft + 1.5 * IQR_sqft))]
print(f"Outliers em pre√ßos: {len(outliers_price)}")
print(f"Outliers em sqft_living: {len(outliers_sqft)}")
```
